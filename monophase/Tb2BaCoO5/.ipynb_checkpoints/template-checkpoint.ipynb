# background

```python
import matplotlib.pyplot as plt
```

```python
bac = pd.read_csv(r'bac.csv',header=None )
upbac = pd.read_csv(r'upbacground.csv',header=None )
```

```python
Var = 18
fig = plt.figure(figsize=[5,3])
plt.subplot()
plt.plot(bac.iloc[:,0],bac.iloc[:,1],c = 'k', linestyle = '-',label = 'background function')
plt.plot(bac.iloc[:,0],bac.iloc[:,1] + Var,c = 'r', linestyle = '--',label = 'confidence interval')
plt.plot(bac.iloc[:,0],bac.iloc[:,1] - Var,c = 'r', linestyle = '--')
plt.scatter(upbac.iloc[:,0],upbac.iloc[:,1],c = 'cyan',s=20, label = 'background intensity')
plt.legend(fontsize = 12)
plt.savefig('.png',dpi=900)
```
